```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import math
from sherpa.astro.ui import *

def bb_fit(spectrum_text,range_=None) :
    load_data(1,spectrum_text,2)
    set_staterror(0.01, fractional=True)
    #plot_data()
    set_source(bbody.bb)
    print(bbody.bb)
    bb.space = 1
    bb.kT.min = 0.0001
    if range_ == None :
        notice()
    else :
        notice(range_)        
    fit()
    plot_fit()
    
    #d1 = get_data(); print d1.y  
    
    print 'Temp is: ', bb.KT.val * 11604*1000, 'K'
    A = bb.ampl.val
    radius = math.sqrt(A/9.884e31) * 3.15e8 # pc
    radius_SI = radius * 3.08567758128E+16
    radius_solar = radius_SI / 6.95e8
    print ' Radius is: ', radius_SI, "metres or ", radius_solar, "suns"

bb_fit('/home/treynolds/data/SN2016gsd/blackbody/20161009/20161009_zap_scaled_dr_z.txt',
        #range_="4350:4500,5000:5650,6850:7570")
        range_="3650:7500")


```
